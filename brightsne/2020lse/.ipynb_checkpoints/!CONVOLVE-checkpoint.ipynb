```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as matlib

from astropy.io import ascii
from astropy.table import QTable, Table, Column, unique
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
from astropy.convolution import convolve, Gaussian1DKernel, Box1DKernel
from astropy.modeling.functional_models import Gaussian1D

from scipy.signal import convolve as scipy_convolve
from scipy.interpolate import interp1d
```

### Combining averaged light curve (1 day MJDbinsize) with simulated Gaussian precursor

```python
lc = ascii.read('2020lse_i000.o.1days.lc.txt')
plt.figure(0)
plt.plot(lc['MJD'],lc['uJy'], 'ro',ms=2)
xlim1=57750
xlim2=59050
plt.xlim(xlim1,xlim2)
plt.title('2020lse light curve')

def combinelcgauss(mean, amplitude=500.0, stddev=100, nsigma=5, xlim1=xlim1, xlim2=xlim2):
    print('Combining light curve with Gaussian with a mean of %s, amplitude of %s, stddev of %s, and nsigma of %s.' % (mean, amplitude, stddev, nsigma))
    s1 = Gaussian1D()
    s1.mean=mean
    s1.amplitude=amplitude
    s1.stddev=stddev
    r = np.arange(s1.mean-5*s1.stddev, s1.mean+5*s1.stddev, s1.stddev/10.0)
    plt.figure(1)
    plt.plot(r, s1(r), 'bo', ms=2)
    plt.axhline(linewidth=1,color='k')
    plt.title('Gaussian distribution')
    
    nsigma=5
    indeces = np.where(np.logical_and(lc['MJD']<=s1.mean+nsigma*s1.stddev, lc['MJD']>=s1.mean-nsigma*s1.stddev))
    lc['uJy'][indeces] += s1(lc['MJD'][indeces])
    
    combx = lc['MJD']
    comby = lc['uJy']
    plt.figure(2)
    plt.plot(combx, comby, 'ko', ms=2)
    plt.xlim(xlim1,xlim2)
    plt.title('Combined Guassian and light curve')
    
    return(combx, comby)

combx, comby = combinelcgauss(mean=58250.0, amplitude=500.0, stddev=100, nsigma=5)
```

### Interpolation and convolution

```python
def convolve_lc(MJD, flux, deltat_days=2, interptype='linear', kerneltype='gaussian', kernelsize_days=[50,100,200], xlim1=xlim1, xlim2=xlim2):
    #print('Convolving light curve with %d deltat_days, %s kernel type, and ',kernelsize_days,' kernelsize_days. Interpolation type is %s.' %(deltat_days, kerneltype, interptype))
    
    minmjd = int(min(MJD)+1)
    maxmjd = max(MJD)
    n = int((maxmjd-minmjd)/deltat_days)
    
    # interpolate MJD and flux
    xnew = np.linspace(minmjd, minmjd+n*deltat_days, num=n+1, endpoint=True)
    f = interp1d(MJD, flux, kind=interptype)
    
    plt.figure(0)
    plt.plot(MJD, flux, 'o',ms=3)
    plt.plot(xnew, f(xnew), '--')
    plt.xlim(xlim1,xlim2)
    plt.legend(['original',interptype])
    plt.title('Interpolated light curve (%s)' % interptype)
        
    if kerneltype is 'gaussian':
        plt.figure(1)
        plt.plot(xnew, f(xnew), 'ro', ms=5, label='Original light curve')
        
        colors=['k','b','g','y','c','m','y','w']
        colorcounter=0
        
        for size in kernelsize_days:
            # calculate stddev
            stddev = size/deltat_days 
            print('kernelsize: ',size)
            print('Gaussian kernel stddev: ', stddev,'\n')
            
            # define kernel and convolve
            gauss_kernel = Gaussian1DKernel(stddev=stddev)
            smoothed_gauss = convolve(f(xnew), gauss_kernel)
            
            plt.plot(xnew, smoothed_gauss, color=colors[colorcounter], ms=3, label='Kernelsize %s' % size, alpha=0.5, linewidth=2)
            colorcounter += 1
            
        plt.xlim(xlim1,xlim2)
        plt.legend(loc='best')
        plt.title('Convolution using Gaussian kernel')
        plt.show()
        
    elif kerneltype is 'box':
            # calculate boxwidth
            boxwidth = kernelsize_days/deltat_days 
            print('Box kernel width: ', boxwidth)

            # define kernel and convolve
            box_kernel = Box1DKernel(width=boxwidth)
            smoothed_box = convolve(f(xnew), box_kernel)

            plt.figure(1)
            plt.plot(xnew, f(xnew), 'ro', ms=5, label='Original light curve')
            plt.plot(xnew, smoothed_box, 'k-', ms=3, label='Box convolved', alpha=0.5, linewidth=2)
            plt.xlim(xlim1,xlim2)
            plt.legend(loc='best')
            plt.title('Convolution using box kernel')
            plt.show()
    
    else: 
        print('Kernel type must be Gaussian or box!')

convolve_lc(combx, comby, kerneltype='gaussian')
```
